```python
import random
import string
import tkinter as tk

# Character set and shuffled set for encryption and decryption
a = " " + string.punctuation + string.ascii_letters + string.digits
a = list(a)
b = a.copy()
random.shuffle(b)

# Function to encrypt the input text
def encrypt_message():
    message = text_input.get("1.0", tk.END).strip()
    encrypted_message = ''
    for char in message:
        if char in a:
            index = a.index(char)
            encrypted_message += b[index]
        else:
            encrypted_message += char
    text_output.delete("1.0", tk.END)
    text_output.insert(tk.END, encrypted_message)

# Function to decrypt the input text
def decrypt_message():
    encrypted_message = text_input.get("1.0", tk.END).strip()
    decrypted_message = ''
    for char in encrypted_message:
        if char in b:
            index = b.index(char)
            decrypted_message += a[index]
        else:
            decrypted_message += char
    text_output.delete("1.0", tk.END)
    text_output.insert(tk.END, decrypted_message)

# GUI setup using Tkinter
root = tk.Tk()
root.title("Encrypt and Decrypt Messages")

frame = tk.Frame(root)
frame.pack(pady=20)

label = tk.Label(frame, text="Enter your message:")
label.grid(row=0, column=0)

text_input = tk.Text(frame, height=5, width=40)
text_input.grid(row=1, column=0)

encrypt_button = tk.Button(frame, text="Encrypt", command=encrypt_message)
encrypt_button.grid(row=2, column=0, sticky="ew", pady=5)

decrypt_button = tk.Button(frame, text="Decrypt", command=decrypt_message)
decrypt_button.grid(row=3, column=0, sticky="ew", pady=5)

text_output = tk.Text(frame, height=5, width=40)
text_output.grid(row=4, column=0)

root.mainloop()
```
